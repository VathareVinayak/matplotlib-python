```python
import matplotlib.pyplot as plt
from random import randint
%matplotlib inline
```

```python
students = np.arange(1,11)
marks = np.random.randint(80,100,10)
print(students,"\n")
print(marks)
```

**SCATTER PLOT using MATPLOTLIB**

```python
plt.scatter(students,marks,c='b')
plt.xlabel('Roll Number of Students')
plt.ylabel('Marks of students')
plt.title('Student data')
plt.savefig('data_config.png')
```

**PLOT BY plt**

```python
plt.plot(students , marks ,'*r',linestyle = 'dashed')
plt.xlabel('roll_no')
plt.ylabel("marks")
plt.title("visualization using plot by plt")
```

**SUBPLOTING using plot**
 

```python
plt.subplot(2,2,1)
plt.plot(students,marks,'r--')

#second plot
plt.subplot(2,2,2)
plt.scatter(students,marks, c='r')

#Third Plot
plt.subplot(2,2,3)
plt.plot(students,marks,'r*')

#Forth plot
plt.subplot(2,2,4)
plt.scatter(students,marks, c='b')
```

**BAR PLOT**

```python
 #BAR PLOT
x_one = [1,3,5]
y_one = [11,12,13]

x_two = [2,4,6]
y_two = [5,14,16]

plt.bar(x_one,y_one)
plt.bar(x_two,y_two,color='g',)
plt.title('checking value')
plt.xlabel('x_axis')
plt.ylabel('y_axis')
plt.show()
```

**HISTOGRAM**

```python
hist_array = np.array([21,25,35,85,87,5,6,65,42,52,62,76])
plt.hist(hist_array,bins=15)
plt.title('histogram')
```

**BOX PLOT**

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

#ploting the box 
plt.boxplot(data , vert= True ,patch_artist=True)
```

**PIE CHART**

```python
labels = 'python','java','R','C++'
size = [150,150,170,130]
colors  = ['gold','yellowgreen','lightskyblue','lightcoral']
explode = (0.2,0,0,0)


#plotting the pie chart  
plt.pie(size,labels=labels,colors=colors , explode= explode ,autopct= '%1.1f%%' , shadow='True')
plt.axis=('equal')
```
